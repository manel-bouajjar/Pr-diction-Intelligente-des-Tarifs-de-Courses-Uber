```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
df = pd.read_csv('fichier_nettoye.csv') 
```

```python
df.head()
```

```python
# Select features (X) and target variable (y)
X = df.drop(columns=['fare_amount']) # Drop target variable and other␣non-feature columns
y = df['fare_amount']

```

```python
# Transformation des caractéristiques catégoriques en variables indicatrices (encodage one-hot)
X = pd.get_dummies(X)
```

```python
df.head()
```

```python
# Division des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Initialisation et entraînement du modèle Random Forest
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
# Prédiction sur l'ensemble de test
y_pred = model.predict(X_test)
```

```python
# Évaluation du modèle
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
mae = mean_absolute_error(y_test, y_pred)
```

```python
 # Validation croisée avec 5 plis
cv_scores = cross_val_score(model, X, y, cv=5, scoring='r2')
mean_r2 = cv_scores.mean()
```

```python
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("Mean Absolute Error (MAE):", mae)
print("Mean R²:", mean_r2)
```
